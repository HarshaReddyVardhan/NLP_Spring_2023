```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.util import ngrams
import docx2txt
import docx
import random
from operator import itemgetter
```

```python
text = docx2txt.process("5 papers related to blockchain.docx")
text = text.lower()
words = word_tokenize(text)
```

```python
from collections import Counter
freq = Counter(words)

# Print each token and its frequency
for token, count in freq.items():
    print(f"{token}: {count}")
```

```python
doc = docx.Document('5 papers related to blockchain.docx')

with open('input.txt', 'w',encoding="utf-8") as f:
    for para in doc.paragraphs:
        f.write(para.text)
        f.write('\n')
```

```python
filename = "input.txt"

with open(filename,encoding="utf-8") as f:
    content = f.readlines()
content = [ line for line in content if line != '\n' ]
len(content)
```

```python
tokens = [t for l in content for t in l.split() ]
# for x in tokens:
#     print(x,end='//')
```

```python
tl = len(tokens)
tl
```

```python
bi_freq = nltk.FreqDist(tokens)
bi_probs = {k: v/tl for (k,v) in bi_freq.items()}
bi_pvals = list(bi_probs.values())
bi_cumprobs = {k: sum(bi_pvals[0:ix+1]) for ix, (k,v) in enumerate(bi_probs.items())}
```

```python
cfd_bigram = nltk.ConditionalFreqDist(nltk.bigrams(tokens))
```

```python
cfd_bigram.items()
```

```python
len(cfd_bigram.items())
```

```python
for j in range(5):
    keys = list(bi_cumprobs.keys())
    values = bi_cumprobs.values()
    sentence = ''
    for i in range(10):
        rn = random.random()
        ix = [i for (i, v) in enumerate(values) if v > rn][1]
        sentence += keys[ix] + ' '
    print('Sentence '+ str(j+1)+' : ')
    print(sentence)
```
